```python
filelist = []
NBFILES=numstop-numstart+1
for index in range(numstart,numstop+1):
    if index != 102:
        filename=root_filename+str(index)+ext_filename
        fullfilename=os.path.join(outpath,filename)
        filelist.append(fullfilename)
```

```python
fig1 = plt.figure(figsize=(20,15))
for index,file in enumerate(filelist):       
    array=np.loadtxt(file)
    x=array[:,0]
    y=array[:,1]
    plt.plot(x,y,'-')   
plt.xlim([0, 1000.])
plt.tick_params(axis='y', which='minor')
plt.grid(True,which='both')
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.xlabel('$\lambda$ (nm)',fontsize=20)
plt.ylabel('air attenuation',fontsize=20)
plt.title('Atmospheric attenuation at NOFS',fontsize=30)
plt.legend(loc='best')
```
