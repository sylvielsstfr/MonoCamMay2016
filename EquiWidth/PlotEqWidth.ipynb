```python
# to enlarge the sizes
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
```

```python
class PDF(object):
  def __init__(self, pdf, size=(200,200)):
    self.pdf = pdf
    self.size = size

  def _repr_html_(self):
    return '<iframe src={0} width={1[0]} height={1[1]}></iframe>'.format(self.pdf, self.size)

  def _repr_latex_(self):
    return r'\includegraphics[width=1.0\textwidth]{{{0}}}'.format(self.pdf)
```

```python
AirMass=np.array( [ 1.7710645458576959, 1.4353960968637178, 1.2256076238203211,1.1070173712070741])
```

```python
dataO2=np.array([ 1.6248595458109285, 1.5533474638378753, 1.5278171326625194,1.4368520791415296])
```

```python
errdataO2=np.array( [ 0.028028537750969685, 0.02039842190470165, 0.02430873507831563,0.030612215520111681])
```

```python
predO2=np.array([ 3.3790377804954916, 3.0626187869218646, 2.8339278751825034,2.6914602158721053])
```

```python
errpredO2=np.array([ 0.034702095234333888, 0.03386344678347164, 0.033419132990500422,0.033197288271909584])
```

```python
plt.figure(figsize=(10,6))
plt.plot(AirMass,dataO2,'r--')
plt.errorbar(AirMass,dataO2,yerr=errdataO2 ,
                                 fmt='s',
                                 markersize = 3,
                                 color = 'r',
                                zorder = 300,
                                antialiased = True,linewidth=2,label='data')
plt.plot(AirMass,predO2,'b--')
plt.errorbar(AirMass,predO2,yerr=errpredO2 ,
                            fmt = 'o',
                            markersize = 3,
                            color = 'b',
                            zorder = 300,
                            antialiased = True,linewidth=2,label='libradtran')
plt.xlabel('airmass')
plt.ylabel('equivalent width in nm')
plt.ylim(0.,5.)
plt.title('equivalent width for atmospheric O$_2$ absorption line (759 nm)')
plt.legend(loc='upper left')
```
