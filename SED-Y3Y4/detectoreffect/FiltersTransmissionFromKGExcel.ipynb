## Comparison Y3/Y4

```python
plt.figure(figsize=(10,6))
plt.plot(wl_y3,flt_y3,color='r',label='Y3')
plt.plot(wl_y4,flt_y4,color='b',label='Y4')
plt.legend(loc='best')
plt.title('Comparison of Y3/Y4 transmission')
plt.xlabel('wavelength in nm')
plt.ylabel('transmission in %')
```
