```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
%matplotlib inline
```

```python
pre_filename=['pwv05_','pwv10_','pwv15_','pwv20_','pwv25_','pwv30_','pwv35_',
              'pwv40_','pwv45_','pwv50_','pwv55_','pwv60_','pwv65_','pwv70_','pwv75_','pwv80_']
```

```python
NBPWV=len(pre_filename)
```

```python
currentpath=os.getcwd()
outpath='./monocam_2016-05-09-y4/afglus'
root_filename='UVSPEC_REPTRAN_SOLAR_ALT22_COARSE_'
ext_filename='.out'
numstart=1
numstop=8
```

```python
list_filelist = []

for idx,pre_f in enumerate(pre_filename):
    filelist = []
    NBFILES=numstop-numstart+1
    for index in range(numstart,numstop+1):
        if index != 102:
            filename=root_filename+pre_filename[idx]+str(index)+ext_filename
            fullfilename=os.path.join(outpath,filename)
            filelist.append(fullfilename)
    list_filelist.append(filelist)
```

```python
f, axarr = plt.subplots(NBPWV, sharex=True,sharey=True,figsize=(20,30))
for idx in range(NBPWV):
    for index in range(NBFILES):
        file=list_filelist[idx][index]
        array=np.loadtxt(file)
        x=array[:,0]
        y=array[:,1]
        axarr[idx].plot(x,y,'-',label=pre_filename[idx]) 
        axarr[idx].grid(True,which='both')
        axarr[idx].set_title(pre_filename[idx],fontsize=20)   
plt.xlim([300, 1200.])
plt.tick_params(axis='y', which='minor')
plt.grid(True,which='both')
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.xlabel('$\lambda$ (nm)',fontsize=20)
#plt.ylabel('air attenuation',fontsize=20)
plt.suptitle('Atmospheric attenuation at NOFS',fontsize=20)
#plt.legend(loc='best')
```
